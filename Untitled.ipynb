```python
import numpy as np
import matplotlib.pyplot as plt
import math

## first-order
def Gillespie(k1,k2,rxn1_order,particle_nb):
    # Define stoichiometry matrix and reaction parameters
    stoichiometry_matrix = np.array([
        [-1,  1,  0],  # Reaction 1: A -> B
        [ 0, -1,  1]   # Reaction 2: B -> C
    ])
    
    # Parameters
    k1 = 1  # [=] 1/sec
    k2 = 1  # [=] 1/sec, Rate constant for B -> C
    
    # Unit conversion
    # Define initial particles first
    initial_counts = [10000, 0, 0]  # [=] particles
    NA = 6.022e23  # Avogadro's number [=] particles/moles
    # Then define concentrations
    initial_concentrations = [1, 0, 0] # [=] moles/L
    # So can determine volume of the system
    volumes = [initial_counts[0]/(NA*initial_concentrations[0]), 0, 0] # [=] L
    
    # Define rate laws
    def rates(concent):  # Input should be a NumPy array
        concent = np.asarray(concent)  # input should be a list of # of particles, NOT CONCENTRATION
        A, B, C = concent
        return np.array([
            k1 * A ** rxn1_order,  # First-order reaction A -> B
            k2 * B   # First-order reaction B -> C
        ])
    
    # Run Gillespie algorithm
    x = np.array(initial_counts, dtype=float) # variable
    t = 0
    times = [t]
    part = [x.copy()] # save history
    
    # initial_rates = rates(initial_concentrations)
    # r_total = initial_rates[0]+initial_rates[1]
    t_max = 5 # set the t_max of the interest so that reducing the computationing time
    while t < t_max :
        # Step 1: Compute reaction probabilities
        # unit conversion
        concent = x/(NA*volumes[0])
        reaction_rates = rates(concent)
        # print(reaction_rates)
        r_total = np.sum(reaction_rates[0]+reaction_rates[1])
        # print(r_total)
        if r_total <= 0:
            break
        
        # Step 2: Generate random numbers and calculate time step
        p1, p2 = np.random.uniform(0, 1, size=2)
        tau = -np.log(p1) / (r_total*NA*volumes[0])
        # why should r_total be devided by NA*volumes[0]:
        # gillespie reaction rate: r1 = k1*x_A [=] particles/sec
        # reaction rate of mine:   r1 = k1*C_A [=] moles/sec
        # so I should convert units so that times are not changing with the number of particles
        
        # Step 3: Determine which reaction occurs
        m = 1 if p2 > reaction_rates[0]/r_total else 0
        
        # Step 4: Update time and species counts
        t += tau
        x += stoichiometry_matrix[m]
        
        times.append(t)
        part.append(x.copy())
    
    # rescaling times
    times = np.array(times)
    part = np.array(part)
    concentrations = part/(NA*volumes[0])
    
    # Plot results
    plt.title(f'X_A = {initial_counts[0]} particles \nk1=k2={k1}/sec \nVolume={volumes[0]:.3e}L')
    plt.plot(times, concentrations[:, 0], label = 'C_A')
    plt.plot(times, concentrations[:, 1], label = 'C_B')
    plt.plot(times, concentrations[:, 2], label = 'C_C')
    # plt.xlim(0, 5)
    plt.xlabel('times [=] sec')
    plt.ylabel('Concentrations [=] M')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

## second order
# Define stoichiometry matrix and reaction parameters
stoichiometry_matrix = np.array([
    [-1,  1,  0],  # Reaction 1: A -> B
    [ 0, -1,  1]   # Reaction 2: B -> C
])

# Parameters
k1 = 1  # [=] 1/(M*sec)
k2 = 1  # [=] 1/sec

# Unit conversion
# Define initial particles first
initial_counts = [10000, 0, 0]  # [=] particles
NA = 6.022e23  # Avogadro's number [=] particles/moles
# Then define concentrations
initial_concentrations = [1, 0, 0] # [=] moles/L
# So can determine volume of the system
volumes = [initial_counts[0]/(NA*initial_concentrations[0]), 0, 0] # [=] L

# Re-define Parameters
k11 = k1/(NA*volumes[0]) ** 2 # [=] moles²/(sec ⋅ particles² ⋅ L²)
k22 = k2/(NA*volumes[0]) # [=] moles/(particles*sec*L)

# Define rate laws
def rates(particles):  # Input should be a NumPy array
    particles = np.asarray(particles)  # input should be a list of # of particles, NOT CONCENTRATION
    A, B, C = particles  
    return np.array([
        k11 * A ** 2,  # First-order reaction A -> B
        k22 * B   # First-order reaction B -> C
    ])

# Run Gillespie algorithm
concentrations = np.array(initial_concentrations, dtype=float) # variable
t = 0
times = [t]
concen = [concentrations.copy()] # save history

initial_rates = rates(initial_counts)
r_total = initial_rates[0]+initial_rates[1]
while r_total > 0 :
    # Step 1: Compute reaction probabilities
    reaction_rates = rates(concentrations*NA*volumes[0])
    # print(reaction_rates)
    r_total = np.sum(reaction_rates[0]+reaction_rates[1])
    # print(r_total)
    if r_total < 0:
        break
    
    # Step 2: Generate random numbers and calculate time step
    p1, p2 = np.random.uniform(0, 1, size=2)
    tau = -np.log(p1) / r_total
    # print(tau)
    
    # Step 3: Determine which reaction occurs
    m = 1 if p2 > reaction_rates[0]/r_total else 0
    
    # Step 4: Update time and species counts
    t += tau
    concentrations += stoichiometry_matrix[m]/(NA*volumes[0])
    
    times.append(t)
    concen.append(concentrations.copy())

# rescaling times
times = np.array(times)/(NA*volumes[0])
concen = np.array(concen)

# Plot results
plt.title(f'The stochastic solution w/ 2nd order reaction only in rxn 1 \nk1={k1}moles²/(sec⋅particles²⋅L²) \nand k2={k2}moles/(sec⋅particles⋅L)')
plt.plot(times, concen[:, 0], label = 'C_A')
plt.plot(times, concen[:, 1], label = 'C_B')
plt.plot(times, concen[:, 2], label = 'C_C')
plt.xlabel('times [=] sec')
plt.ylabel('Concentrations [=] M')
plt.xlim(0, 5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
C_A0 = 1.0  # Initial concentration (M)
k = 1.0     # Rate constant (1/s)
t_values = np.linspace(0, 5, 100)  # Time range

# Compute concentrations for 1st order reaction
C_A_first_order = C_A0 * np.exp(-k * t_values)

# Compute concentrations for 2nd order reaction
C_A_second_order = C_A0 / (1 + k * C_A0 * t_values)

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(t_values, C_A_first_order, label="1st Order: $C_A = C_{A0} e^{-kt}$", linestyle='dashed', color='blue')
plt.plot(t_values, C_A_second_order, label="2nd Order: $C_A = \\frac{C_{A0}}{1 + k C_{A0} t}$", linestyle='solid', color='red')

# Labels and legend
plt.xlabel("Time (s)")
plt.ylabel("Concentration of A (M)")
plt.title("Comparison of 1st and 2nd Order Reactions: $A \\to B$")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
