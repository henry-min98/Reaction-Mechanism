```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import math
```

# Case 1: Both reactions are first-order

# 1. Numeric Solution (ODE)

```python
C_A0 = 1.0  # Initial concentration of A (mol/L)
C_B0 = 0.0  # Initial concentration of B (mol/L)
C_C0 = 0.0  # Initial concentration of C (mol/L)
k1 = 1    # Rate constant for A -> B (1/s)
k2 = 1    # Rate constant for B -> C (1/s)
vol = 1 # [=] L
vol_flow = 0.5 # [=] L/sec
tau = vol/vol_flow   # [=] sec, Space time (residence time) in seconds. We have constant tau (constant volumetric flow)

def cstr_odes(t, C):
    C_A, C_B, C_C = C  
    dC_A_dt = (C_A0 - C_A) / tau - k1 * C_A
    dC_B_dt = (C_B0 - C_B) / tau + k1 * C_A - k2 * C_B
    dC_C_dt = (C_C0 - C_C) / tau + k2 * C_B
    return [dC_A_dt, dC_B_dt, dC_C_dt]

t_span = (0, 20)  
t_eval = np.linspace(0, 20, 100)  

sol = solve_ivp(cstr_odes, t_span, [C_A0, C_B0, C_C0], t_eval=t_eval, method='RK45')
t_values = sol.t
C_A_values, C_B_values, C_C_values = sol.y

# Plot results
plt.plot(t_values, C_A_values, label=r'$C_A$', linewidth=2)
plt.plot(t_values, C_B_values, label=r'$C_B$', linewidth=2)
plt.plot(t_values, C_C_values, label=r'$C_C$', linewidth=2)
plt.xlabel('Time [=] sec',fontsize=18)
plt.ylabel('Concentration [=] M',fontsize=18)
plt.title(f"k1={k1}sec⁻¹, k2={k2}sec⁻¹, τ={tau}sec",fontsize=18)
plt.legend()
plt.show()
```

```python
# Steady-state solution
C_A_ss = C_A0 / (1 + k1 * tau)
C_B_ss = (k1 * C_A_ss) / (k2 + 1/tau)
C_C_ss = C_A0 - C_A_ss - C_B_ss

print(f"Steady-state concentrations:")
print(f"C_A = {C_A_ss:.4f} mol/L")
print(f"C_B = {C_B_ss:.4f} mol/L")
print(f"C_C = {C_C_ss:.4f} mol/L")
print('These values are corresponding to the values in the plots below at the same space time')
```

# 2. Analytic Solution

```python
CA0 = 1.0

# Space time (τ) range for calculation
tau = np.linspace(0.1, 10, 500)  # Space time range (s)
# we are varying space time, which means with the same volume, the volumetric flow (vol/time) is decreasing as time goes pass

parameter_sets = [
    (10, 1),  # Case 1: k1 >> k2
    (1, 1),   # Case 2: k1 == k2
    (1, 10)   # Case 3: k1 << k2
]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
cases = ["k1 >> k2", "k1 = k2", "k1 << k2"]

for i, (k1, k2) in enumerate(parameter_sets):
    CA = CA0 / (1 + k1 * tau)
    CB = (CA0 * k1 * tau) / ((1 + k1 * tau) * (1 + k2 * tau))
    CC = CA0 - CA - CB

    ax = axes[i]
    ax.plot(tau, CA, label="C_A")
    ax.plot(tau, CB, label="C_B")
    ax.plot(tau, CC, label="C_C")
    ax.set_xlabel("Space Time (τ) [=] sec",fontsize=18)
    ax.set_title(f"Case: {cases[i]} (k1={k1}sec⁻¹, k2={k2}sec⁻¹)",fontsize=18)
    if i == 0:
        ax.set_ylabel("Concentration [=] M",fontsize=18)
    ax.set_xlim(0,5)
    ax.set_ylim(0,1)
    ax.legend()

plt.tight_layout()
plt.show()
print('As Space time increases, the volumetric flow (vol/time) decreases')
```

# Case 2: Second order reaction in rxn 1

# 1. Numeric Solution

```python
C_A0 = 1.0  # Initial concentration of A (mol/L)
C_B0 = 0.0  # Initial concentration of B (mol/L)
C_C0 = 0.0  # Initial concentration of C (mol/L)
k1 = 1    # Rate constant for A -> B (1/s)
k2 = 1    # Rate constant for B -> C (1/s)
vol = 1 # [=] L
vol_flow = 0.5 # [=] L/sec
tau = vol/vol_flow   # [=] sec, Space time (residence time) in seconds. We have constant tau (constant volumetric flow)

def cstr_odes(t, C):
    C_A, C_B, C_C = C  
    dC_A_dt = (C_A0 - C_A) / tau - k1 * C_A ** 2
    dC_B_dt = (C_B0 - C_B) / tau + k1 * C_A ** 2 - k2 * C_B
    dC_C_dt = (C_C0 - C_C) / tau + k2 * C_B
    return [dC_A_dt, dC_B_dt, dC_C_dt]

t_span = (0, 20)  
t_eval = np.linspace(0, 20, 100)  

sol = solve_ivp(cstr_odes, t_span, [C_A0, C_B0, C_C0], t_eval=t_eval, method='RK45')
t_values = sol.t
C_A_values, C_B_values, C_C_values = sol.y

# Plot results
plt.plot(t_values, C_A_values, label=r'$C_A$', linewidth=2)
plt.plot(t_values, C_B_values, label=r'$C_B$', linewidth=2)
plt.plot(t_values, C_C_values, label=r'$C_C$', linewidth=2)
plt.xlabel('Time [=] sec',fontsize=18)
plt.ylabel('Concentration [=] M',fontsize=18)
plt.title(f"k1={k1}(Msec)⁻¹, k2={k2}sec⁻¹, τ={tau}sec",fontsize=18)
plt.legend()
plt.ylim(0,1)
plt.show()
```

```python
# Steady-state solution
C_A_ss = (-1+math.sqrt(1+4*C_A0*k1*tau))/(2*k1*tau)
C_B_ss = k1*tau*C_A_ss**2/(k2*tau+1)
C_C_ss = k2*tau*C_B_ss

print(f"Steady-state concentrations:")
print(f"C_A = {C_A_ss:.4f} mol/L")
print(f"C_B = {C_B_ss:.4f} mol/L")
print(f"C_C = {C_C_ss:.4f} mol/L")
print('These values are corresponding to the values in the plots below at the same space time')
```
